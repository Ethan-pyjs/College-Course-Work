### 4. What are the first 25 top-rated movies with 100 or more reviews?
- We want the movies that have highest average of rating.
- Also, those movies with the highest number of reviews.
- We will use the movie and the ratings datasets to find the answer.
- We use the Pandas merge() and groupby():
    - merge() to merge the datasets.
    - groupby() to group the variables.


Steps:
1. merge the datasets and name it merged_df.
2. groupby() the "movie title", and "rating", and name it rating_count.
3. get the average of all the ratings for each movie, and name it avg_rating.
4. filter the movies with more than 100 reviews, and name it movie_100_reviews.
5. use the isin() to check inside both avg_rating, and movie_100_reviews variables to get the movies that answer the question.




1. Merging two datasets
2. Grouping the datasets

```python
movie.head()
```

```python
ratings.head()
```

```python
ratings
```

```python
ratings.shape
```

```python
# 1. merging the two datasets
merged_df = movie.merge(ratings, on='movie id')
```

```python
merged_df.head(20)
```

```python
merged_df.tail()
```

```python
merged_df.shape
```

```python
# 2. groupby() the "movie title", and "rating"
# and name it rating_count
rating_count = merged_df.groupby(['movie title'])['rating'].count().reset_index()
rating_count.head(10)
```

```python
# 3. get the average of all the ratings for each movie
# and name it avg_rating
avg_rating = merged_df.groupby(['movie title'])['rating'].mean().sort_values(ascending=False).reset_index()
avg_rating.head()
```

```python
# 4. filter the movies with more than 100 reviews
# and name it movie_100_reviews
movie_100_reviews = rating_count[rating_count['rating']>100]['movie title']
movie_100_reviews.head()
```

```python
avg_rating[avg_rating['movie title'].isin(movie_100_reviews)]
```

```python
top_25_movies = avg_rating[avg_rating['movie title'].isin(movie_100_reviews)].head(25)
top_25_movies
```

```python
len(top_25_movies)
```

### What is the relationship between movie genres and gender?

```python
all_data = merged_df.merge(user, on='user id')
```

```python
genre_by_gender = all_data.groupby('gender').sum().loc[:,'Action':'Western']
```

```python
genre_by_gender['total'] = all_data['gender'].value_counts()
```

```python
genre_by_gender.T
```

```python
(genre_by_gender.div(genre_by_gender.total, axis=0)* 100).T
```

## Conclusion

List 15 insights here:
1. Adding T to a dataframe will transpose it, meaning that the rows and columns will be swapped. This is helpful to make a dataframe easier to look at if the dataframe happens to have more rows than columns or more columns than rows.
2. The sum() returns a sum, if used in a dataframe, it returns the sum values of each variable(columns).
3. Using the sort_values() method will sort the dataframe, this is set default to sort ascending.
4. For sort_values() method to sort descending, it has to have ascending set to False. Which would be: ascending=False. 
5. The loc() is used to index and search for certain values in the dataframe's variables. This is helpful for looking for specific values in the dataframe.
6. Setting axis to 1 will guide a function to just the columns of a dataframe. On the other hand, setting axis to 0 will guide a function to just the rows of a dataframe. By default, axis is always set to 0.
7. The merge() method will merge datasets together using a common variable. This is helpful when dealing with multiple sheets of datasets. 
8. The groupby() method will group variables in a dataframe. This can be done to create a new, smaller dataframe from the original dataframe. This is helpful for merged dataframes since merged dataframes tend to become extremely large after being merged. 
9. Using the value_counts() method will show the total value count of a variable in the dataframe.
10. Using isin() method will check if the dataframe has certain values. 
11. Using the reset_index() method will reset the index of a dataframe and give it a new, default index. 
12. Using mean() will return the average of the values.
13. When using head(), the amount of rows can be specified by adding a value into head(), like head(10). This would return the first 10 rows of the dataframe rather than 5.
14. According to the data mining on the relationship of gender and movie genres, males significantly review more movies than females do.
